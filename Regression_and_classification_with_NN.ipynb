```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# **Regression**

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Salary_Data.csv')
df.head()
```

```python
df.dropna(inplace = True)
```

```python
x = df[['Age', 'Years of Experience']]
y = df.Salary
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=5)
```

```python
scalar = StandardScaler()
xtrain_s = scalar.fit_transform(xtrain)
```

```python
smodel = Sequential([
    Input(shape = (2,)),
    Dense(4, activation= 'relu'),
    Dense(2, activation= 'relu'),
    Dense(1)
])
```

```python
smodel.compile(optimizer='sgd', loss='mae')
```

```python
history = smodel.fit(xtrain_s, ytrain, epochs= 70)
```

```python
xtest_s = scalar.transform(xtest)
```

```python
loss = smodel.evaluate(xtest_s, ytest)
print(loss)
```

```python
# Predict
xnew = np.array([[46,10]])
```

```python
xnew_s = scalar.transform(xnew)
```

```python
sal_pred = smodel.predict(xnew_s)
```

```python
sal_pred
```

# **Classification**

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Diabetes.csv')
df.head()
```

```python
df.dropna(inplace = True)
```

```python
x = df[['Age', 'Glucose']]
y = df.Diabetes
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=5)
```

```python
scalar = StandardScaler()
xtrain_s = scalar.fit_transform(xtrain)
```

```python
cls_model = Sequential([
    Input(shape=(2,)),
    Dense(8, activation= 'relu'),
    Dense(4, activation = 'relu'),
    Dense(1, activation = 'sigmoid')
])
```

```python
cls_model.compile(optimizer='sgd', loss='binary_crossentropy', metrics= ['accuracy'])
```

```python
history = cls_model.fit(xtrain_s, ytrain, epochs= 70)
```

```python
# Evaluate
xtest_s = scalar.transform(xtest)
```

```python
loss = cls_model.evaluate(xtest_s, ytest)
print(loss)
```

```python
# Predict
xnew = np.array([[46,100]])
```

```python
xnew_s = scalar.transform(xnew)
```

```python
prediction = cls_model.predict(xnew_s)
```

```python
prediction
```

```python
if (prediction < 0.5):
  result = ['negative']
else:
  result = ['positive']
```
